### Distribuições



```python
name = 'Este é um texto que será usado para análise em um gráfico de barras.'
frequency = {}
for c in name:
    frequency[c] = frequency.get(c, 0) + 1
    
print('Freq1:', frequency)

freq2 = collections.Counter(name) # método alternativo para contar frequências
print('Freq2:', freq2)

#plt.bar(frequency.keys(), frequency.values())
plt.bar(freq2.keys(), freq2.values())
plt.show()
```
