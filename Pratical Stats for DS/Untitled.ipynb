```python
import numpy as np
import math
```

```python
dataset = [1,1,2,3,4,5,5,5,5,6,7,8,9]
md = [1,2,3,4,5,6]

populacao = [1,2,1,4,3,2,4,1,5,2,3,4,2,5,1,3,2,4,1,4,2,2,4,1,3,4,5,1]
amostra = [1,5,2,3,4,2,5,1,3,2,4,1,4]
```

```python
def media(ds):
    return sum(ds)/len(ds)

media(dataset)
```

```python
def mediana(ds):
    ds.sort()
    if len(ds) % 2 != 0:
        return ds[int(len(ds)/2)]
    else:
        pos = int(len(ds)/2)-1
        return (ds[pos] + ds[pos+1])/2

mediana(md)
```

```python
def moda(ds):
    ds.sort()
    freq = dict()
    tem_moda = False # booleano para checar se existe moda, ou seja, 
                    # algum valor se repete mais que uma Ãºnica vez
    maior_freq = 0
    for x in ds:
        if x in freq:
            freq[x] += 1
            tem_moda = True # existe moda nesse dataset
            if freq[x] > maior_freq:
                maior_freq = x
        else:
            freq[x] = 1
    if tem_moda:
        return ds[maior_freq]
        
moda(dataset)
```

```python
def variancia_populacao(ds):
    md = media(ds)
    total = 0
    for x in ds:
        total += (x - md)**2
    return total/len(ds)

def variancia_amostra(ds):
    md = media(ds)
    total = 0
    for x in ds:
        total += (x - md)**2
    return total/(len(ds)-1)

var_pop = variancia_populacao(populacao)
var_ams = variancia_amostra(amostra)

print('var_populacao: ', var_pop, ' var_amostra: ', var_ams)
```
